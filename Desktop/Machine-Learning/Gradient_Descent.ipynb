```python
#Gradient discent for f(x)=x**2
```

```python
def gradient_descent_1d(start_x,learning_rate,iterations):
    x=start_x

    for i in range(iterations):
        gradient=2*x
        x=x-learning_rate*gradient

        print(f"Interation{i+1}: x={x:.6f},f(x)={x*x:.6f}")
    return x

gradient_descent_1d(start_x=10,learning_rate=0.1,iterations=10)
```

```python
#Gradient Descent for f(x,y)=x**2+y**2
```

```python
def gradient_descent_2d(start_x,start_y,learning_rate,iterations):
    x=start_x
    y=start_y

    for i in range (iterations):
        grad_x=2*x
        grad_y=2*y

        x=x-learning_rate*grad_x
        y=y-learning_rate*grad_y

        print(f"Interations {i+1}:x={x:.6f},y={y:.6f},f(x,y)={x*x+y*y:.6f}")

    return x,y

gradient_descent_2d(start_x=8,start_y=6,learning_rate=0.1,iterations=10)
        
```
